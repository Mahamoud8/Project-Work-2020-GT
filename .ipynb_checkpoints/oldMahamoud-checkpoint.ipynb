```python
def elises_func(stringy):
        our_float = stringy.replace("   ","").replace("  ","").replace(" ","").replace(",","").replace("$","").replace("-","")
        if not our_float:
            our_float = 0
        else:
            our_float = float(our_float)
        
        return our_float
#    return float(stringy.replace(",","").replace("$","").replace(" ","").replace("-","").replace("","0"))
```

```python
elises_func("149,998.30")
```

```python
#PPP_data.info()
```

```python
PPP_data["Loan Amount"] = PPP_data["Loan Amount"].apply(elises_func)
PPP_data.head()
```

```python
# Slice the data and place it into bins
pd.cut(PPP_data["Loan Amount"], bins, labels=group_labels).head()
```

```python
# Place the data series into a new column inside of the DataFrame
PPP_data["View Loan Amount"] = pd.cut(PPP_data["Loan Amount"], bins, labels=group_labels)
PPP_data.head()
```

```python
# Create a GroupBy object based upon "View Group"
zip_group = PPP_data.groupby("View Loan Amount")

# Find how many rows fall into each bin
print(zip_group["Loan Amount"].count())

# Get the average of each column within the GroupBy object
zip_group[["Loan Amount", "City"]].mean()
```
